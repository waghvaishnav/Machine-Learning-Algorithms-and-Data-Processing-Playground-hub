```python
from sklearn.linear_model import SGDRegressor
import pandas as pd
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

data = {
    "area": [850, 900, 1100, 1200, 1500, 1600, 1700, 1800, 2000, 2100, 2300, 2500, 2700, 3000, 3200],
    "price": [45, 48, 60, 65, 80, 85, 90, 95, 110, 115, 125, 135, 150, 165, 180]
}

df = pd.DataFrame(data)
df.head()

x=df[["area"]]
y=df["price"]

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2)


model = SGDRegressor(max_iter=1000,eta0= 0.01,learning_rate="constant",random_state=42)
model.fit(x_train,y_train)

result = model.predict(x_test)

print(result)

print("coeficient",model.coef_)
print("intercept is ",model.intercept_)

print("mse of the gradient model is : ",mean_squared_error(y_test,result))
```
