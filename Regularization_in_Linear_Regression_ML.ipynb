```python
import numpy as np
from sklearn.linear_model import LinearRegression,Ridge,Lasso

x = np.array([[1], [2], [3], [4], [5]])
y = np.array([1.1, 2.0, 2.9, 4.1, 5.0])

#linear regression
lin = LinearRegression().fit(x, y)

# ridge regression
ridge = Ridge(alpha=1.0)
ridge.fit(x,y)
# lasso regression
lasso = Lasso(alpha=0.1)
lasso.fit(x,y)

print("Linear Coeff:", lin.coef_)
print("Ridge Coeff:", ridge.coef_)
print("Lasso Coeff:", lasso.coef_)
```

```python
#prediction using the linear,lasso,ridge :
print("lasso result :",lasso.predict([[10]]))
print("ridge result :",ridge.predict([[10]]))
print("linear simple model prediction :",lin.predict([[10]]))
```
