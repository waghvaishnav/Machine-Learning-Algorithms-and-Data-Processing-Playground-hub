```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB,GaussianNB
from sklearn.metrics import classification_report
data = pd.read_csv("spam.csv")

df = pd.DataFrame(data)

print(df.sample(5))

print(df['Category'].value_counts())

x_train,x_test,y_train,y_test = train_test_split(df.Message,df.Category,test_size=0.2,random_state=42)
cv = CountVectorizer()

x_train_vect = cv.fit_transform(x_train)
x_test_cv = cv.transform(x_test)

x_test_cv_arr = x_test_cv.toarray()
print(x_test_cv_arr.shape)

mnb_model = MultinomialNB()
gnb_model = GaussianNB()


```

## Multinomial Model Training  and Evaluation for Text classification:

```python
mnb_model.fit(x_train_vect,y_train)
y_pred_mnb = mnb_model.predict(x_test_cv)
print(classification_report(y_test,y_pred_mnb))
```

## GaussianNB Model Training and Evaluation for Text classification:

```python
gnb_model.fit(x_train_vect.toarray(),y_train)
y_pred_gnb= gnb_model.predict(x_test_cv.toarray())
print(classification_report(y_test,y_pred_gnb))
```
