# project 1 :

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import numpy as np
from sklearn.metrics import mean_absolute_error,mean_squared_error

# from google.colab import files
# files.upload()  # Choose the file from your computer

df = pd.read_csv("study_hours_marks.csv")
print(df.head())

x = df[['study_hours']]     #double breckets 2d inputs.
y = df['marks']

model = LinearRegression()
model.fit(x,y)



predicted_marks = model.predict(x)


mae =mean_absolute_error(y,predicted_marks)
mse =mean_squared_error(y,predicted_marks)
rmse = np.sqrt(mse)

print("mae",mae)
print("mse",mse)
print("rmse",rmse)


new_prediction = model.predict([[17]])[0]
print("marks prediction for the 17 hours is to be ",new_prediction)
```

```python
data=pd.read_csv("Students_Grading_Dataset_Biased.csv")
data.head(10)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
import numpy as np

model = LinearRegression()

x = data[['Study_Hours_per_Week']]
y = data['Final_Score']

model.fit(x,y)

predicted_score = model.predict(x)

#valid regression metrics :
MAE  =mean_absolute_error(y,predicted_score)
MSE  =mean_squared_error(y,predicted_score)
RMSE =np.sqrt(MSE)
R2   =r2_score(y,predicted_score)

print("MAE:",round(MAE,2))
print("MSE:",round(MSE,2))
print("RMSE:",round(RMSE,2))
print("R2:",round(R2,2))


#HISTOGRAMS :
plt.figure(figsize=(10,6))
plt.hist(data['Final_Score'],bins=30,color="skyblue",edgecolor="black")
plt.grid(True)
plt.xlabel("student scores")
plt.ylabel("number of students")
plt.title("student score distribution")
plt.tight_layout()

plt.show()

plt.figure(figsize=(10,6))
plt.scatter(x,y,marker="*",color="orange",label="student actual scores")
plt.plot(x,predicted_score,color="red",label="predicted_score(regression_line)")
plt.legend()
plt.grid(True)
plt.xlabel("study hours")
plt.ylabel("final output")
plt.title("model prediction vs actual score")

plt.tight_layout()
plt.show()


new_prediction = model.predict([[9]])[0]
print("model prediction for the 9 hours the score is to be  : ",round(new_prediction,2))
```
