Feature scaling : "train_test_split".

```python
from sklearn.model_selection import train_test_split

df=pd.DataFrame({"study_hours":[1,2,3,4,5,6,7],
                 "marks":[23,54,67,87,98,90,98]})

x=df["study_hours"]
y=df["marks"]


x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state= 1 )

print("x_train : ")
print(x_train,"\n")
print("x_test : ")
print(x_test,"\n")
print("y_train")
print(y_train,"\n")
print("y_test")
print(y_test,"\n")

```
