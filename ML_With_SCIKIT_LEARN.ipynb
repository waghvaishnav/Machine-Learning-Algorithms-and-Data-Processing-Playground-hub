#Linear_regression

```python
from sklearn.linear_model import LinearRegression

study_hours = [[1],[3],[5],[7],[9],[10],[28]]
marks =  [12,23,45,65,75,85,100]

hours= int(input("Enter the number of hours you studied : "))

reg_model=LinearRegression()

reg_model.fit(study_hours,marks)

prediction = reg_model.predict([[hours]])

print(f"your result for the regression model for the {hours} hr. the marks is to be {prediction} .")
```

#Logistic regression

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()

hours = [[1],[2],[3],[4],[5],[6],[7]]
marks= [0,0,0,1,1,1,1]  # 0 fail,1 pass .

value = float(input("Enter the hours you studied : "))
model.fit(hours,marks)

result = model.predict([[value]])

if result == 1:
  print(f"you studied {value} hr. result is PASS . ")
else:
  print(f"you studied {value} hr. result is FAIL . ")
```

#K-Nearest Neighbors classifier:

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors=3)

x = [[100,20],       #fruits weight,size in cm.
     [120,27],
     [135,31],
      [138,34],
     [140,40],
     [160,45]]

y = [0,0,0,0,1,1]         # fruit type apple or orange.

model.fit(x,y)

weight=int(input("Enter the weight of the fruit : "))
size = int(input("Enter the size of the fruit : "))

pred = model.predict([[weight,size]])[0]        # The [0] is there because model.predict() returns a NumPy array, even if it predicts only one value.

if pred == 0:
  print("the predicted fruit is the apple")
else:
  print("the predicted fruit is the orange.")
print(pred)
```

# Decision Tree Algorithm :

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt

model=DecisionTreeClassifier()
x = [                           # Features: [weight (grams), size (cm)]
    [100, 20],          #apple
    [120, 25],               #apple
    [130, 28],            #apple
    [140, 35],            #orange
    [150, 40],
    [170, 45]
]

y = [0, 0, 0, 1, 1, 1]                 # Labels: 0 = Apple, 1 = Orange

model.fit(x,y)

size=int(input("Enter the size of the fruit : "))
shade = int(input("Enter the shade of the fruit : "))

result=model.predict([[size,shade]])[0]

if result == 0:
  print("This is likely an apple.")
else:
  print("This is likely an Orange.")

plt.figure(figsize=(10,16))
tree.plot_tree(model,feature_names=["Weight", "Size"], class_names=["Apple", "Orange"], filled=True)
plt.show()
```
